```python
from pytorch_mums.mums import MUMS
import yaml
import torch.utils.data as data
from torch import min, max
```

```python
with open('config.yaml') as f:
    config = yaml.load(f, Loader=yaml.FullLoader)

# use instrument_family and instrument_source as classification targets
dataset = MUMS(root=config['path']['mums'],
    include_dirs=config['include_dirs'],
    blacklist_pattern=config['blacklist_pattern'],
    categorical_field_list=['instrument_family_str', 'instrument_source_str'])

# NB: audio samples are not fixed length
loader = data.DataLoader(dataset, batch_size=1, shuffle=True)
# yield batch from loader
for samples, instrument_family_target, instrument_source_target, targets \
        in loader:
    print(samples.shape, instrument_family_target.shape,
            instrument_source_target.shape)
    print(min(samples), max(samples))
```
